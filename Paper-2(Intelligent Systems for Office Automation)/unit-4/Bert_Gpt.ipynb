```python
from transformers import pipeline
classifier=pipeline("text-classification")
result=classifier("I love this movie!")
print(result)
```

```python
from transformers import pipeline
classifier=pipeline("text-classification", model="akilesh96/autonlp-mrcooper_text_classification-529614927")
result=classifier("Congratulation! you won a free iphone. click here to claim.")
print(result)
```

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="savasy/bert-base-turkish-sentiment-cased")
result = classifier("Congratulation! you won a free iphone. click here to claim.")
print(result)
```

```python
from transformers import pipeline

classifier = pipeline("text-classification", model="savasy/bert-turkish-text-classification")
result = classifier("I hate this product, it is a complete waste of money.")
print(result)
```

```python
qa_pipeline=pipeline("question-answering")
context="I am living in Madurai"
question="Where i was living?"
print(qa_pipeline(question=question,context=context))
```

```python
generator=pipeline("text-generation", model="gpt2")
print(generator("Arificial intelligence in transforming",max_new_tokens=10))
```

```python
summarization=pipeline("summarization", model="facebook/bart-large-cnn")
text="Artificial intelligence is transforming industrie ..."
print(summarization(text, max_length=50, min_length=25))
```
